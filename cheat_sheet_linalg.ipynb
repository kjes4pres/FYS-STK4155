# **Matrix & Vector Cheat Sheet for Least-Squares / Derivatives**

### **1. Matrix multiplication**

* Dimensions must match: $(n \times m)(m \times p) = n \times p$
* Usually **not commutative**: $\boldsymbol{A}\boldsymbol{B} \neq \boldsymbol{B}\boldsymbol{A}$
* Transpose rule:

$$
(\boldsymbol{A}\boldsymbol{B})^T = \boldsymbol{B}^T \boldsymbol{A}^T
$$

---

### **2. Scalars**

* Scalars are $1 \times 1$ matrices:

$$
(\boldsymbol{x}^T \boldsymbol{A} \boldsymbol{s})^T = \boldsymbol{s}^T \boldsymbol{A}^T \boldsymbol{x} = \boldsymbol{x}^T \boldsymbol{A} \boldsymbol{s}
$$

* Useful for combining terms in expansions.

---

### **3. Derivatives (matrix calculus)**

**Linear term**: $\boldsymbol{s}^T \boldsymbol{b}$

$$
\frac{\partial}{\partial \boldsymbol{s}} (\boldsymbol{s}^T \boldsymbol{b}) = \boldsymbol{b}^T
$$

**Quadratic term**: $\boldsymbol{s}^T \boldsymbol{M} \boldsymbol{s}$

* If $\boldsymbol{M}$ is symmetric:

$$
\frac{\partial}{\partial \boldsymbol{s}} (\boldsymbol{s}^T \boldsymbol{M} \boldsymbol{s}) = 2 \boldsymbol{s}^T \boldsymbol{M}
$$

**Norm squared**: $\|\boldsymbol{x}-\boldsymbol{A}\boldsymbol{s}\|^2 = (\boldsymbol{x}-\boldsymbol{A}\boldsymbol{s})^T(\boldsymbol{x}-\boldsymbol{A}\boldsymbol{s})$

$$
\frac{\partial}{\partial \boldsymbol{s}} \|\boldsymbol{x}-\boldsymbol{A}\boldsymbol{s}\|^2 = -2 (\boldsymbol{x}-\boldsymbol{A}\boldsymbol{s})^T \boldsymbol{A}
$$

---

### **4. Symmetry**

* $\boldsymbol{A}^T \boldsymbol{A}$ is always symmetric: $(\boldsymbol{A}^T \boldsymbol{A})^T = \boldsymbol{A}^T \boldsymbol{A}$
* Simplifies quadratic derivatives.

---

### **5. Factoring derivatives**

* Example:

$$
2 \boldsymbol{s}^T \boldsymbol{A}^T \boldsymbol{A} - 2 \boldsymbol{x}^T \boldsymbol{A} = -2 (\boldsymbol{x} - \boldsymbol{A}\boldsymbol{s})^T \boldsymbol{A}
$$

* Keep row vs column orientation consistent: derivative of scalar w\.r.t column vector → row vector.

---

### **6. Quick tips**

* Always **check dimensions**.
* Scalars can be transposed freely.
* For least-squares: the derivative always involves the **residual** $(\boldsymbol{x}-\boldsymbol{A}\boldsymbol{s})$.
* Remember: linear → derivative = coefficients, quadratic → derivative = $2 \times$ matrix times vector.
